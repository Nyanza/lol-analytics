```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

from selenium.webdriver.support.ui import Select, WebDriverWait
driver = webdriver.Chrome()
driver.get("https://champion.gg/statistics/")
select = driver.find_element_by_class_name("navigation-rank-selector")  #get the select element            
options = select.find_elements_by_tag_name("option") #get all the options into a list

optionsList = []

for option in options: #iterate over the options, place attribute value in list
    optionsList.append(option.get_attribute("value"))

for optionValue in optionsList:
    print("starting loop on option", optionValue)

    select = Select(driver.find_element_by_class_name("navigation-rank-selector"))
    select.select_by_value(optionValue)
    soup = BeautifulSoup(driver.page_source,'lxml')
    table = soup.find_all('table')[0] 
    df = pd.read_html(str(table))[0]
    filename = optionValue + '.csv'
    df.to_csv(filename, index=False, encoding='utf-8') 
```

```python
import requests
import os.path
import urllib.request


url = 'http://localhost:8000/api/champions'
g = requests.get(url)
#get sprite data
path = './assets/sprite/'
for champ in g.json():
    print(champ)
    # check if img exists in ./assets/sprite
    if not os.path.isfile(path + champ['img']):
        # if not, download
        urllib.request.urlretrieve("http://ddragon.leagueoflegends.com/cdn/7.5.2/img/sprite/" + champ['img'], path + champ['img'])
```

```python
url = 'http://localhost:8000/api/champions'
g = requests.get(url)
#get champion image data
path = './assets/champion/'
for champ in g.json():
    print(champ)
    fullChamp = requests.get(url+'/'+champ['name']).json()[0]
    # check if img exists in ./assets/sprite
    if not os.path.isfile(path + fullChamp['img']):
        # if not, download
        try:
            urllib.request.urlretrieve("http://ddragon.leagueoflegends.com/cdn/7.5.2/img/champion/" + fullChamp['img'], path + fullChamp['img'])
        except:
            continue
```

```python
url = 'http://localhost:8000/api/champions'
g = requests.get(url)
#get champion image data
path = './assets/loading/'
img_url = 'http://ddragon.leagueoflegends.com/cdn/img/champion/loading/' #key_skinNum.jpg
for champ in g.json():
    fullChamp = requests.get(url+'/'+champ['name']).json()[0]
    # if not, download
    key = fullChamp['key']
    for skin in fullChamp['skins']:
        ext = key + '_' + str(skin['num'])  + '.jpg'
        print(img_url + ext)
        # check if img exists in ./assets/sprite
        if not os.path.isfile(path + ext):
            try:
                urllib.request.urlretrieve(img_url + ext, path + ext)
            except:
                continue
```
